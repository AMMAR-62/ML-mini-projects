```python
import cv2 as cv
import mediapipe as mp
import numpy as np
import time
```

```python
mpDraw = mp.solutions.drawing_utils
mpPose = mp.solutions.pose
pose = mpPose.Pose()
```

```python
video = cv.VideoCapture("./static/2.mp4")

pTime = 0
if not video.isOpened():
    print("can't read the video")
    exit(0)

while (True):
    ret , frame = video.read()
    cTime = time.time()
    fps = 1/(cTime-pTime)
    ptime = cTime
    if ret:
        # keep the aspect ratio
        # scale_percent = 30
        # width = int(frame.shape[1] * scale_percent / 100)
        # height = int(frame.shape[0] * scale_percent / 100)
        # dim = (width, height)

        # do not keep the aspect ratio.
        width = 1000
        height = 700
        dim = (width, height)
        frame = cv.resize(frame, dim)

        # changing color
        frameRGB = cv.cvtColor(frame, cv.COLOR_BGR2RGB)
        
        # processing the model
        results = pose.process(frameRGB)

        # if landmarks are detected
        if results.pose_landmarks:
            mpDraw.draw_landmarks(frame, results.pose_landmarks, mpPose.POSE_CONNECTIONS)
            for id, lm in enumerate(results.pose_landmarks.landmark):
                print(id, lm)
                cx, cy = int(lm.x*width), int(lm.y*height)
                cv.circle(frame, (cx, cy), 10, (255, 255, 255), cv.FILLED)

        cv.putText(frame, str(int(fps)), (70,50), cv.FONT_HERSHEY_PLAIN, 3, (255,255,255), 3)
        cv.imshow("video1", frame)
        if cv.waitKey(25) & 0xFF == ord('q'):
            break
    else:
        break

video.release()
cv.destroyAllWindows()
    
```
